####rank: 특정 열의 값에 따라 순위를 매겨주는 함수

df[열].rank(method = ..., ascending = True/False)   True: 오름차순 / False: 내림차순

####[method]  
- average: 동점 관측치 간의 그룹 내 평균 순위 부여 
- min: 동점 관측치 그룹 내 최소 순위 부여
- max: 동점 관측치 그룹 내 최대 순위 부여
- first: 동점 관측치 중 데이터 상에서 먼저 나타나는 관측치부터 순위 부여
- dense: 최소값과 같으나 그룹간 순위가 1씩 증가 

<3등이 2명일 경우>  
- average: 둘 다 3.5등 -> 4등 없고 -> 5등
- min: 둘 다 3등 -> 4등 없고 -> 5등
- max: 둘 다 4등 -> 5등 (3등 없음)
- first: 먼저 발견된 3등이 3등 -> 그 다음 4등 -> 5등
- dense: 둘 다 3등 -> 4등  (min 방법과 동일하지만 다음 등수에 차이가 있음)

```python
df = pd.read_excel("성적 처리_데이터.xlsx", sheet_name = 'Sheet1')
df
```

```python
df['총합'] = df.iloc[:,2:].sum(axis = 1)
df
```

```python
df['평균'] = df['총합']/5
df
```

```python
df['순위'] = df['평균'].rank(method='average', ascending=False)
df
```

```python
df['순위'] = df['평균'].rank(method='min', ascending=False)
df
```

```python
df['순위'] = df['평균'].rank(method='max', ascending=False)
df
```

```python
df['순위'] = df['평균'].rank(method='first', ascending=False)
df
```

```python
df['순위'] = df['평균'].rank(method='dense', ascending=False)
df
```
