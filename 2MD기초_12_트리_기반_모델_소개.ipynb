#### 결정트리
- 스무고개 하듯이 예/아니오 질문을 이어가며 학습하는 모델
- 첫 질문을 Root Node라고 하며, 뒤에 질문은 Node라고 표현. 
- 가장 마지막 노드를 Leaf Node라고 부름
- 연산을 하지 않으므로 데이터를 표준화 할 필요가 없다. 

#### 결정트리 학습 과정
- 가장 잘 데이터를 구분할 수 있는 질문으로 기준을 나눈다. 
- 그 뒤에 또 다시 데이터를 가장 잘 구문할 수 있는 기준을 나누고, 이 과정을 계속 반복한다. 
- 하지만 지나치게 많이 데이터를 나누면 과대적합이 일어날 수 있다.

#### 랜덤 포레스트
- 결정트리는 학습에서 과대적합되는 경향이 있음. 
- 랜덤 포레스트는 다수의 결정트리로부터 분류 결과를 취합해서 결론을 얻는 방법

```python
# 랜덤 포레스트
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
data, target = iris['data'], iris['target']
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier(max_depth=3)  # max_depth=3: 3층에서 멈춰라 / 과대적합을 막을 수 있다. 
dt.fit(train_input, train_target)
dt.score(test_input, test_target)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10))

plot_tree(dt, feature_names = iris['feature_names'], filled=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()  # 대부분 랜덤트리가 결정트리보다는 무조건 좋게 나옴. 
rf.fit(train_input, train_target)
rf.score(test_input, test_target)
```
