#### concat: 여러 개의 데이터프레임을 하나의 데이터프레임으로 합치는 함수
- 열 이름을 기준으로 합치기: pd.concat([df1, df2, ...])
- 행 이름을 기준으로 합치기: pd.concat([df1, df2, ...], axis = 1)    
- axis = 1일 때: 열 방향(좌→우)으로 행 이름을 기준으로 붙이기 
- axis = 0일 때: 행 방향(위→아래)으로 열 이름을 기준으로 붙이기
- 행 이름 바꾸기: df.index = [...]
- 열 이름 바꾸기: df.columns = [...]

```python
import pandas as pd
```

```python
df1 = pd.read_csv('concat_1.csv')
df2 = pd.read_csv('concat_2.csv')
df3 = pd.read_csv('concat_3.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
concat1 = pd.concat([df1, df2, df3])    # axis = 0이 기본값
print(concat1)
```

```python
concat1 = pd.concat([df1, df2, df3], ignore_index=True)   # 행번호 정렬
print(concat1)
```

```python
concat2 = pd.concat([df1, df2, df3], axis = 1)
print(concat2)
```

```python
concat2['A']
```

```python
concat2 = pd.concat([df1, df2, df3], axis = 1, ignore_index=True)  # 열이름 정렬
print(concat2)
```

```python
# 열이름이 다른 데이터프레임이 있다면 합쳐줄까?

df2.index
```

```python
df2.index = [0,3,5,7]
df2
```

```python
pd.concat([df1, df2, df3], axis = 1)   # 행번호가 달라도 강제로 NaN 값을 채워서 합쳐줌.
```

```python
df2.columns = ['A', 'C', 'E', 'F' ]
df2
```

```python
pd.concat([df1, df2, df3], axis = 0) # 열 이름이 달라도 강제로 NaN 값을 채워서 합쳐줌.
```
