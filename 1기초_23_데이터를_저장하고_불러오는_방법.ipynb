```python
# CSV파일 불러오기
pd.read_csv('파일이름.csv')
pd.read_csv('파일이름.csv', encoding='euc-kr', sep=",") 

매개변수
encoding: 문자표현방식 / 한글이 포함된 경우 인코딩방식 설정. (유니코드, cp949, euc-kr 등)
sep: 데이터를 읽어올 때 구분자
```

```python
# TSV파일 불러오기
pd.read_csv('파일이름.tsv', encoding='euc-kr', sep="\t")  # 탭으로 구분

매개변수
encoding: 문자표현방식 
sep: 데이터를 읽어올 때 구분자
```
