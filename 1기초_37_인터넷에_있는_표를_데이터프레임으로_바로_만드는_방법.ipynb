#### 표 수집
- 수집한 표 정보를 바로 표(데이터프레임)으로 변환할 수 있다. 


```
pd.read_html(str(HTML)  # 문자열로 바꿔줘야 함. 
```



````python
# 네이버 -> 증권 -> 국내증시 -> 시가총액 -> 코스피&코스닥을 하나의 데이터프레임으로 만드는 방법
````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
````

````python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1')
html = BeautifulSoup(url.text)
kospi_page = int(html.find('td', class_ = "pgRR").find('a')['href'].split('=')[-1])

total = []

for n in range(1, kospi_page + 1):  # 1~38페이지까지 반복

    url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={}'.format(n))
    html = BeautifulSoup(url.text)

    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(html))[1]    # 하나의 리스트로 출력되므로 리스트 안에 인덱스를 지정하여 데이터를 꺼내오기,  
    table = table[table['종목명'].notnull()]      # 실선으로 표시된 부분까지 불러와서 NaN 값이 보이므로, '종목명'에 NaN값이 있는 행 제거

    del table['N']       # 필요없는 열 삭제
    del table['토론실']

    total.append(table)
    time.sleep(1)   # 트래픽 발생을 최소화하기 위해

    print("{}페이지 저장완료...".format(len(total)))
````

````python
len(total)
````

````python
kospi = pd.concat(total, ignore_index = True)
kospi.to_excel("kospi.xlsx")
kospi
````

````python
# 1. html.find('td', class_ = "pgRR"): 태그명 'td', class 속성값 'pgRR'인 것 찾기
html.find('td', class_ = "pgRR")
````

````python
# 2. 찾은 결과에서 다시 find('a')로 a를 찾기
html.find('td', class_ = "pgRR").find('a')
````

````python
# 3. 찾은 결과 'a'에서 'href'에 접근하기
html.find('td', class_ = "pgRR").find('a')['href']
````

````python
# 4. 원하는 결과값 38을 얻기 위해 split('=') 함수를 이용하여 '='을 기준으로 나누기
html.find('td', class_ = "pgRR").find('a')['href'].split('=')
````

````python
# 5. 결과값에서 '38'은 리스트 내 맨 마지막에 있으므로 인덱스 -1로 지정하여 추출
html.find('td', class_ = "pgRR").find('a')['href'].split('=')[-1]
````

````python
# 6. 문자열 -> 숫자열로 바꿔 kospi_page변수에 저장
kospi_page = int(html.find('td', class_ = "pgRR").find('a')['href'].split('=')[-1])
````

````python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1')
html = BeautifulSoup(url.text)
kosdaq_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1])

total2 = []

for n in range(1, kosdaq_page + 1):
    url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={}'.format(n))
    html = BeautifulSoup(url.text)

    table = html.find('table', class_ = 'type2')
    table = pd.read_html(str(html))[1]
    table = table[table['종목명'].notnull()]

    del table['N']
    del table['토론실']

    total2.append(table)
    time.sleep(1)
    print("{}페이지 저장완료...".format(len(total2)))

````
