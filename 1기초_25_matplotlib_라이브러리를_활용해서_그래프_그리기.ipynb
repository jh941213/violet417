```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
fig = plt.figure()    # 그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당

ax1 = fig.add_subplot(1, 1, 1)    # 1행 1열 1번째 자리
ax1.hist(tips['total_bill'], bins=10)      # 히스토그램
```

```python
ax1.hist(tips['total_bill'], bins=20)
fig
```

```python
ax1.hist(tips['total_bill'], bins=30)
fig
```

```python
fig = plt.figure()    
ax1 = fig.add_subplot(1, 1, 1) 

ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size'])  # c는 size마다 색상을 다르게 표시
ax1.set_xlabel('TOTAL BILL')
ax1.set_ylabel('TIP')
ax1.set_title('SCATTER')
```

```python
fig = plt.figure()    
ax1 = fig.add_subplot(1, 1, 1) 

ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size'], s=tips['size'] * 20)  # s는 동그라미 크기 조정 / size의 갭차이를 만들기 위해 20을 곱함. 
ax1.set_xlabel('TOTAL BILL')
ax1.set_ylabel('TIP')
ax1.set_title('SCATTER')
```

```python
fig = plt.figure()    
ax1 = fig.add_subplot(1, 1, 1) 

ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size'], s=tips['size'] * 20, alpha=0.5)  # alpha는 투명도 설정 
ax1.set_xlabel('TOTAL BILL')
ax1.set_ylabel('TIP')
ax1.set_title('SCATTER')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)

male = tips[tips['sex'] == 'Male']    #tips에 접근해서 tips의 성별이 Male인 것만 뽑아 male 변수에 넣어줌
female = tips[tips['sex'] == 'Female']    #tips에 접근해서 tips의 성별이 Female인 것만 뽑아 female 변수에 넣어줌
ax1.boxplot([male['total_bill'], female['total_bill']])    # 상자 그래프
```
