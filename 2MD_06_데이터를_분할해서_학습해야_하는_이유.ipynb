#### 훈련데이터와 시험데이터
- 훈련데이터로 학습하고, 훈련데이터로 평가한다면 정확도를 신뢰할 수 없음. 
- 훈련데이터와 시험데이터를 만들고, 훈련데이터로만 학습하고, 시험데이터로 평가

#### 데이터의 비율
- 최대한 훈련데이터에 많은 비율을 할당하는 것이 좋음
- 데이터가 적다면 최소 75:25 비율을 유지해주는 것이 좋음
- 데이터가 많다면 99:1 또는 99.9:0.1 수준으로 최대한 훈련데이터에 많은 비율을 할당하는 것이 좋음  
=> 최대한 컴퓨터가 많은 데이터를 학습해야만 모르는 무언가의 패턴을 찾아낼 수 있다. 

#### 데이터 분할하기
from sklearn.model_selection import train_test_split  
<br>
- 순서가 중요!!  
훈련문제집, 시험문제집, 훈련정답지, 시험정답지 = train_test_split(문제집, 정답지, test_size = ?)  
ex) test_size = 0.2  8:2로 분할
훈련문제집 80% - 시험문제집 20%
훈련정답지 80% - 시험정답지 20%



```python
import pandas as pd

fish = pd.read_csv('fish.csv')
fish
```

```python
data = fish[['length', 'weight']].to_numpy()    # 2차원 행렬 모양으로 바로 가져오기
data
```

```python
target = fish['class'].to_numpy()
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

train_input, test_input, train_target, test_target = train_test_split(data, target) 
# test_size를 입력 안할 경우, 기본값 = 75:25 비율 | 랜덤으로 분할

knn = KNeighborsClassifier()
knn.fit(train_input, train_target)    # 훈련세트로만 훈련
```

```python
knn.score(test_input, test_target)    # 시험세트로 평가
```

```python
test_target    # 실제값
```

```python
knn.predict(test_input)    # 예측값
```
