#### 데이터 크롤링: 인터넷에 있는 데이터를 가져오는 방법
- 인터넷은 큰 틀로 보면, HTML이라는 언어로 구성되어 있다. 
- 따라서 Python으로 HTML을 다루어야 하며, 페이지 요청을 할 수 있어야 한다.  
  
[라이브러리]
1.   requests: 페이지 요청 라이브러리
2.   bs4의 BeautifulSoup: Python으로 HTML을 다루는 라이브러리
3.   time: 안전하게 수집하는 방법
4.   pandas: 수집한 데이터 엑셀화

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

# requests 활용
- **requests.get('링크')**를 하면 페이지를 요청할 수 있음.  -->  문자열로 된 HTML을 가져옴.

```python
url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90")
# 200: 문제없음 / 400번대: 존재하지 않는 페이지 / 500번대: 존재는 하지만 볼 수 없는 페이지(로그인처리, 잠금, 비공개 처리) 
# response [200]: 데이터를 가져오는데 큰 문제는 없었다. 충돌x
url
```

```python
url.text    # 가져온 내용에 대한 정보를 볼 수 있음. (문자열)
```

# 데이터 크롤링
- **BeautifulSoup(url.text)**를 하면 수집한 데이터를 HTML로 변환할 수 있음.  (문자열을 진짜 HTML로 변환해 줌) 

- find: 해당 HTML을 찾아서 HTML로 반환 --> find(태그, 속성)
- find_all: 해당 HTML을 모두 찾아서 리스트로 반환

```python
html = BeautifulSoup(url.text)
html
```

```python
# 원하는 글자 위에 마우스 오른쪽 버튼 클릭 -> 검사 -> 태그명, 속성 확인 
html.find('span', class_ = 'f_red')    # 태그명: 'span', class 속성값: 'f_red'  <태그명, 속성값> text </태그명>
```

```python
html.find('span', class_ = 'f_red').text    # 중간에 text만 추출
```

```python
current = html.find('span', class_ = 'f_red').text.replace("가", "")    # 추출한 글자의 수정
current
```

```python
url2 = requests.get("https://www.dhlottery.co.kr/gameResult.do?method=byWin")
url2
```

```python
url2.text
```

```python
html2 = BeautifulSoup(url2.text)
html2
```

```python
lotto = html2.find('div', class_ = "nums")
lotto
```

```python
lotto = html2.find('div', class_ = "nums").find_all('span')
lotto
```

```python
box = []

for i in lotto:
    box.append(int(i.text))

box
```
