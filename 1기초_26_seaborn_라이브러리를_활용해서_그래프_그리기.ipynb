```python
# 산점도 / 이변량 그래프
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# 산점도 + 히스토그램 / 이변량 그래프
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# 평균 막대 그래프 / 이변량 그래프
sns.barplot(x='time', y='total_bill', data=tips)
```

```python
# 상자 그래프 / 이변량 그래프
sns.boxplot(x='time', y='total_bill', data=tips)
# outlier 이상치(점) - upper fence: 1.5를 곱한 값. 통계적으로 신뢰할 수 있는 구간(충분히 일어날 수 있는 일) - 상위 25% - median 중앙값 - 상위 75%
```

```python
# 바이올린 그래프 / 데이터가 어느 곳에 밀집되어 있는지 알 수 있지만, 간결하고 눈에 확 띄지 않아 boxplot을 대체하기 어려움. / 이변량 그래프
sns.violinplot(x='time', y='total_bill', data=tips)
```

```python
# 관계 그래프 / 비교할 수 있는 모든 열들을 비교해서 표현해 줌
sns.pairplot(tips)
```

```python
# 다변량 그래프 그리기 / 삼변량 그래프
매개변수 hue: 색상 추가
매개변수 col: 집단별로 그래프 그리기
```

```python
# 집단별로 그래프 그리기 / 삼변량 그래프
FacetGrid: 집단별로 그래프 그리는 함수
sns.set_style(배경): 그래프 배경 설정

darkgrid, whitegrid, dark, white, ticks 등
```

```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'], tips['tip'], shade=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
sns.violinplot(x='sex', y='total_bill', data=tips)
```

```python
sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
anscombe = sns.load_dataset('anscombe')
sns.lmplot(x='x', y='y', data=anscombe)
```

```python
anscombe = sns.load_dataset('anscombe')
sns.lmplot(x='x', y='y', data=anscombe, col='dataset')
```

```python
sns.lmplot(x='x', y='y', data=anscombe, col='dataset', col_wrap = 2)    # 최대 2열까지만 표시 
```

```python
import matplotlib.pyplot as plt

sns.set_style('whitegrid')    # 배경설정

facet = sns.FacetGrid(tips, col='time', hue='sex', row='smoker')    # col=time 열은 시간 집단 / hue=sex 성별 색상 다르게 / row=smoker 행은 흡연테이블 집단 
facet.map(plt.scatter, 'total_bill', 'tip')
```
