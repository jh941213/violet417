```python
# melt: 지정한 열의 데이터를 모두 행으로 내려주는 기능
pd.melt(df, id_vars = ...)

id_vars: 고정할 열    # 절대 행으로 내리면 안되는 열
var_name: 위치를 변경한 열의 이름
value_name: 위치를 변경한 열의 데이터를 저장한 열의 이름
```

```python
# 열 분할하기
# 분할: 시리즈.str.split(기준), 수집: 시리즈.str[인덱스]
```

```python
import pandas as pd
df = pd.read_csv('pew.csv')
df2 = pd.melt(df, id_vars = ["religion"], var_name = 'income', value_name = 'count')
df2
```

```python
billboard = pd.read_csv('billboard.csv')
billboard2 = pd.melt(billboard, id_vars = ['year', 'artist', 'track', 'time', 'date.entered'],
                     var_name = 'week', value_name = 'rating')
billboard2
```

```python
ebola = pd.read_csv('country_timeseries.csv')
ebola2 = pd.melt(ebola, id_vars = ['Date', 'Day'])
ebola2
```

```python
ebola_split = ebola2['variable'].str.split('_')
ebola_split
```

```python
state = ebola_split.str[0]
country = ebola_split.str[1]

ebola2['State'] = state
ebola2['Country'] = country

del ebola2['variable']
ebola2
```
